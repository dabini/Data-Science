#### 이상치 정제

```python
mydf = mpg_df.filter(['drv', 'hwy', 'cty'])
mydf.boxplot()
pyplot.show()
```

```python
third = mydf.quantile(q=0.75)
third
```

```python
first = mydf.quantile(q=0.25)
first
```

```python
# iqr = 3사분위와 1사분위 수의 차
uqr = third - first
```

```python
uqr
```

### 최저 한계치(lower fence) 얻기
최처 한계치는 실제로 존재하는 값이 아닌 계산을 통해 얻어진 수치값이다.

```python
low_fence = first -1.5 *uqr
low_fence
```

### 최고 한계치(upper fence) 얻기
최고 한계치는 실제로 존재하는 값이 아닌 계산을 통해 얻어진 수치값이다.

```python
upper_fence = third + 1.5 * uqr
upper_fence
```

### 아래 수평선 (극단치 경계) 값 얻기
최저한계치보다 큰 녀석들 중에 최소값

```python
mydf[mydf > low_fence].min()
```

### 제일 위 수평선 (극단치 경계) 값 얻기
최고한계치보다 작은 녀석들 중에 최대값

```python
limit = mydf[mydf < upper_fence].max()
```

### 이상치를 결측치로 변경하기
a) 고속도로 연비의 이상치 데이터 추출하기

```python
outlier_df1 = mpg_df.query('hwy>'+  str(limit["hwy"]))
outlier_df1
```

b) 도시 연비의 이상치 데이터 추출하기


```python
outlier_df2 = mpg_df.query('cty >' + str(limit["cty"]))
outlier_df2
```

정제를 위한 복사본 생성

```python
outlier_define_df = mpg_df.copy()
outlier_define_df
```

```python
import numpy
```

```python
for i in outlier_df1.index:
    outlier_define_df.loc[i, 'hwy'] = numpy.nan

outlier_define_df
```

```python
for i in outlier_df2.index:
    outlier_define_df.loc[i, 'cty'] = numpy.nan

outlier_define_df
```

```python
outlier_define_df.filter(['cty', 'hwy']).isna().sum()
```

```python
fuel_mean_df = outlier_define_df.filter(['drv', 'hwy', 'cty']).dropna().groupby('drv').mean()
fuel_mean_df
```

```python
fuel_mean_df.plot.bar(rot=0)
pyplot.grid()
pyplot.xticks([0, 1, 2], ['4륜구동', '전륜구동', '후륜구동'])
pyplot.title(['구동방식별 고속도로, 도시 평균 연비 비교'])
pyplot.ylabel('평균연비')
pyplot.xlable('구동방식')
pyplot.show()
```
